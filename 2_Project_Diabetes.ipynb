```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns


```

```python
#look to see if there is any missing data, Nothing is missing;
df.isnull().sum()
```

```python


#clean data
print(f'original shape of data {df.shape}' )
df=df.drop_duplicates()
print(f'new shape of data {df.shape}' )
df.head()
```

```python



#Using list(df) to get the column headers as a list
column_headers = list(df.columns)
print(column_headers)
```

```python




# dropping passed columns
#df=reassignment instead of inplace. *note- # then df.head
df.drop(["discharge_disposition_id", "admission_source_id","payer_code", "medical_specialty","num_procedures", "num_medications", "number_outpatient","number_emergency", "number_inpatient", "diag_1", "diag_2", "diag_3", "number_diagnoses", "max_glu_serum", "A1Cresult", "metformin", "repaglinide", "nateglinide", "chlorpropamide","glimepiride", "acetohexamide", "glipizide", "glyburide", "tolbutamide", "pioglitazone","rosiglitazone", "acarbose", "miglitol", "troglitazone", "tolazamide","examide", "citoglipton", "insulin", "glyburide-metformin", "glipizide-metformin", "glimepiride-pioglitazone", "metformin-rosiglitazone","metformin-pioglitazone", "change", "diabetesMed", "readmitted"], axis = 1, inplace = True)
df.head()
```

```python




df.tail()
```

```python



#chaining.
#convert bin into a number
#parse the string and get first number and convert to int

df['age_number']= (df["age"].str.replace("[",'')
                  .str.split('-')
                  .str.get(0)
                  .astype(int))
```

```python




#Time in hospital:  2. Number of lab procedure
#df2=df[df['race'].isin(['Caucasian','AfricanAmerican', 'Asian', 'Hispanic', 'Other'])& (df['gender']=='Female')]


ethnicities= ['Caucasian','AfricanAmerican', 'Asian', 'Hispanic', 'Other']
df2=df.query('race in @ethnicities & gender=="Female" & age_number>=60')
df2
```

```python




#second attempt at dropping 'weight'
df2.drop(["weight"],axis = 1, inplace = True)
df2.head()
```

```python
df2.tail()
```

```python





fig,axes=plt.subplots(2,1,figsize=(4,6))
axes=np.ravel(axes)



df2['race'].value_counts().plot(kind='barh',ax=axes[0],xlabel='Number of Patients', alpha=0.5)
df2['age_number'].value_counts().plot(kind='barh',ax=axes[1], xlabel='Number of Patients', ylabel='Age', alpha=0.5)
sns.despine()

```

```python






sns.scatterplot(data=df2, x='time_in_hospital', y= 'num_lab_procedures', hue='race')
```

```python






ax=sns.boxplot(data=df2, x='time_in_hospital', y= 'num_lab_procedures', hue='race')
ax.set(xlabel='Time in Hospital ( Days)', ylabel=' Number of Lab Procedures')
sns.despine()
```
